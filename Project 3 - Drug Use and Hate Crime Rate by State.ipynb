```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Importing Dataset
df = pd.read_excel(r'C:\Users\Jared\Desktop\Homework\Comp Bio + Biostats\Project3Data.xlsx')
df
```

```python
#Creating seaborn histogram of the Drug Use variable, separated by state.
sns.set(style='darkgrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [8, 5]})
sns.displot(df['Drug Use']).set(title='Histogram of Drug Use')

#Creating new dataframe consisting of just the State and Drug Use variables, sorted by Drug Use descending.
drugUse = df.sort_values(by='Drug Use', ascending = False).filter(['State', 'Drug Use'])
drugUse.head()
```

```python
#Reporting the mean Drug Use across all fifty observations.
drugUse.agg('mean')
```

```python
#Creating a histogram of the frequency of Above Average and Below Average Hate Crime Rate States
sns.displot(df['Hateful']).set(title='Histogram of Hate Crime Per Capita by State')
```

```python
#Creating new dataframe that consists of only the variables State and Hateful
hateCrime = df.filter(['State', 'Hateful'])
#Reporting the counts of each category of the Hateful variable
hateCrime.groupby('Hateful').agg('count')
```

```python
#Creating a pie chart that represents the percentage of Drug Use that is found in each category of Hateful.
df.groupby("Hateful")["Drug Use"].sum().plot.pie(autopct="%.1f%%", title= "Drug Use Grouped By Hate Crime Rate")
```
